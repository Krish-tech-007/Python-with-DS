### Data Visualization with Seaborn

Seaborn is a Python Visualization library based on Matplotlib that provides a high level interface for drawing attractive and informative statistical graphs.    
Seaborn helps in creating complex visualization with just a few lines of code.  

```python
import seaborn as sns
```

```python
## Basic plotting with Seaborn
tips = sns.load_dataset('tips') #It has lot of inbuilt datasets
tips
```

```python
## Create a scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
## Line Plot
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line Plot Size vs Total Bill")
plt.show()
```

```python
## Categorical Plots

## Bar Plot
sns.barplot(x='day',y='total_bill',data=tips)
plt.title("Bar Plot Day vs Total Bill")
plt.show()
```

```python
### Box Plot
## A box plot, also known as a box-and-whisker plot, is a graphical representation of a dataset's distribution. 
## It provides a summary of key statistics, including the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum.
## Key Features:
## - Box: Represents the interquartile range (IQR), which contains the middle 50% of the data.
## - Whiskers: Extend from the box to show variability outside the upper and lower quartiles.
## - Median Line: A line inside the box indicating the dataset's midpoint.
## - Outliers: Individual points beyond the whiskers, representing extreme values.
## Box plots are useful for comparing multiple datasets, identifying skewness, and detecting outliers

sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Box Plot Day vs Total Bill")
plt.show()
```

```python
### Violin Plot
## A violin plot is a type of data visualization that combines aspects of a box plot and a density plot to show the distribution of numerical data. 
## It provides more information than a traditional box plot by displaying the probability density of the data at different values, usually smoothed using kernel density estimation (KDE)
## Key Features:
## - Shape: The plot resembles a violin, with symmetrical density curves on either side.
## - Box Plot Elements: It includes markers for the median, interquartile range, and sometimes individual data points.
## - Density Representation: The width of the plot at different values indicates the frequency of data points in that range.
## - Comparison: Useful for comparing distributions across multiple categories.
## Violin plots are particularly helpful when analyzing multimodal data (data with multiple peaks), as they reveal nuances that a box plot might miss. 
## However, they are less commonly used because they can be harder to interpret for those unfamiliar with them.


sns.violinplot(x='day',y='total_bill',data=tips)
plt.title("Violin Plot Day vs Total Bill")
plt.show()
```

```python
### Historgrams
sns.histplot(tips['total_bill'],bins=10,kde=True) #kde stands for kernel density estimator
# It is kept True so that we get the shape of the histogram
# If we keep it False, then we won't get the curve which shows how the data is distributed
```

```python
## KDE plot
sns.kdeplot(tips['total_bill'],fill=True) 
# If fill is kept False, we will only see the curve and the area under the curve won't be shaded.
```

```python
## Pair plot
## To observe the relationship between two variables
## Categorical variables are not taken into consideration
## We need to specifically have a numerical variable     
sns.pairplot(tips)
```

```python
## Heat map
corr = tips[['total_bill','tip','size']].corr() #Correlation
corr
## Interpretation
## If total_bill increases by one unit, tip will increase by 0.675734 units
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
import pandas as pd
sales_df = pd.read_csv('sales.csv')
sales_df.head()
```

```python
## Plot total sales by product
plt.figure(figsize=(10,6))
sns.barplot(x='Product Category', y='Total Revenue',data=sales_df,estimator=sum) 
plt.title("Bar Plot Total Sales by Product")
plt.show()
## Estimator is used to apply aggregator function
```

```python
## Plot total sales by region
plt.plot(figsize=(10,6))
sns.barplot(x='Region',y='Total Revenue',estimator=sum,data=sales_df)
plt.title("Bar Plot Region vs Total Revenue")
plt.show()
```
